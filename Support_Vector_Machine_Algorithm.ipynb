# Iris DataSet

```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df["target"]=iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==2].head()
```

```python
df["flower_name"]=df.target.apply(lambda x:iris.target_names[x])
```

```python
df.head(150)
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
plt.scatter(df0["sepal length (cm)"],df0["sepal width (cm)"],color="CYAN",marker="+")
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],color="green",marker="*")
plt.scatter(df2["sepal length (cm)"],df2["sepal width (cm)"],color="red",marker=".")
plt.title("iris dataset")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.show()
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(["target","flower_name"],axis="columns")
y=df.target
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

# Social Network ads


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
df.head(20)
```

```python
X=df.drop(["Purchased"],axis="columns")
y=df["Purchased"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
print(y_train)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the SVM Model on the Training Set

```python
from sklearn.svm import SVC
classifier=SVC(kernel="linear",random_state=0)
classifier.fit(X_train,y_train)
```

## Prediction

```python
print(classifier.predict(sc.transform([[21,327000]])))
```

```python
print(classifier.predict(sc.transform([[21,3270000]])))
```

```python
print(classifier.predict(sc.transform([[11,327000]])))
```

```python
print(classifier.predict(sc.transform([[1,327000]])))
```

### Prediction on test results

```python
y_pred=classifier.predict(X_test)
```

```python

# Convert the Pandas Series to a NumPy array
y_test_array = y_test.to_numpy()

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test_array.reshape(len(y_test_array),1)),1))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
